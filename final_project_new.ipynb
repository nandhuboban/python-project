```python

print("Processing 'month' column: filling NaNs with 0 and converting to integer...")
df['month'] = df['month'].fillna(0).astype(int)
print("Missing values after 'month' processing:")
display(df.isnull().sum())
```

```python

duplicates_count = df.duplicated().sum()
print(f"\nNumber of duplicate rows found: {duplicates_count}")

if duplicates_count > 0:
    print("Removing duplicate rows...")
    df.drop_duplicates(inplace=True)
    print(f"Number of rows after removing duplicates: {len(df)}")
else:
    print("No duplicate rows found.")
```

```python

print('\nData types after handling month and duplicates:')
display(df.info())
```

```python

for col in ['candidate_type', 'sex', 'party']:
    if df[col].isnull().any():
        mode_val = df[col].mode()[0]
        df[col].fillna(mode_val, inplace=True)
        print(f"Filled missing values in '{col}' with mode: {mode_val}")


for col in ['vote_share_percentage', 'margin_percentage']:
    if df[col].isnull().any():
        mean_val = df[col].mean()
        df[col].fillna(mean_val, inplace=True)
        print(f"Filled missing values in '{col}' with mean: {mean_val:.2f}")

print("\nMissing values after handling all columns:")
display(df.isnull().sum())
```

```python

print('\nData types after handling all missing values:')
display(df.info())
```

```python

df['is_winner'] = df['position'] == 1
print("Created 'is_winner' column. Here's a sample:")
display(df[['candidate_name', 'position', 'is_winner']].head())


df['total_votes_polled'] = df['valid_votes']
print("\nCreated 'total_votes_polled' column. Here's a sample:")
display(df[['pc_name', 'valid_votes', 'total_votes_polled']].head())
```

```python
#Find the distribution of votes


df['vote_share_percentage'] = pd.to_numeric(df['vote_share_percentage'], errors='coerce')
df['vote_share_percentage'].fillna(df['vote_share_percentage'].mean(), inplace=True)
sns.histplot(df['vote_share_percentage'], bins=20)
plt.title('Distribution of Votes')
plt.xlabel('Vote Share Percentage')
plt.ylabel('Frequency')
plt.show()
```

```python


from matplotlib.lines import lineStyles
df_2009_winners=df[(df['position']==1) & (df['year']==2009)]
party_counts_2009=df_2009_winners['party'].value_counts().head(10)
plt.plot(party_counts_2009.index,party_counts_2009.values,marker='o')
plt.title('TOP 10 PARTIES IN loksabha ELECTION')
plt.xlabel('PARTY')
plt.ylabel('NUMBER OF WINNERS')
plt.grid(True)
lineStyles='--'
plt.show()
```

```python
import plotly.express as px

ten_parties_names = df['party'].value_counts().head(10).index

# Filter the DataFrame to include only the top 10 parties
df_top_parties = df[df['party'].isin(ten_parties_names)]
region_sales = df_top_parties.groupby('party')['position'].sum().reset_index()
fig = px.bar(region_sales,
             x='party',
             y='position',
             color='party',
             title='Total Position by Top 10 Parties')

fig.show()
```

```python
from matplotlib.lines import lineStyles
df_2014_winners=df[(df['position']==1) & (df['year']==2014)]
party_counts_2014=df_2014_winners['party'].value_counts().head(10)
sns.barplot(x=party_counts_2014.index, y=party_counts_2014.values, palette=sns.color_palette("tab10", 10))
plt.title('TOP 10 PARTIES IN 2014 losabha ELECTION')
plt.xlabel('PARTY')
plt.ylabel('NUMBER OF WINNERS')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df_2009_winners=df[(df['position']==1) & (df['year']==2009)]
party_counts_2009=df_2009_winners['party'].value_counts().head(10)
sns.boxplot(x=party_counts_2009.index, y=party_counts_2009.values, palette=sns.color_palette("tab10", 10))
```

```python
#multivarie Analysis

top_10_party_names = df['party'].value_counts().head(10).index.tolist()
df_filtered_parties = df[df['party'].isin(top_10_party_names)].copy()

df_filtered_parties['vote_share_percentage'] = pd.to_numeric(df_filtered_parties['vote_share_percentage'], errors='coerce')
df_filtered_parties['vote_share_percentage'].fillna(df_filtered_parties['vote_share_percentage'].mean(), inplace=True)

df_filtered_parties['position_cat'] = df_filtered_parties['position'].astype('category')

plt.figure(figsize=(12, 6))
sns.barplot(data=df_filtered_parties, x='party', y='vote_share_percentage', hue='position_cat', errorbar=None)
plt.title('Average Vote Share Percentage by Top Parties and Position')
plt.xlabel('Party')
plt.ylabel('Average Vote Share Percentage')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Position')
plt.tight_layout()
plt.show()
```

```python
# Create a pivot table to summarize average vote share percentage

pivot_table_avg_vote_share = df_filtered_parties.pivot_table(
    index='party',
    columns='position_cat',
    values='vote_share_percentage',
    aggfunc='mean'
)

print("Pivot Table of Average Vote Share Percentage:")
display(pivot_table_avg_vote_share.head())
```
